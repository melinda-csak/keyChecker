```python
import pandas
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
#fájl elérési helyét, nevét megadni, sep: szeparátor, header: hanyadik sorban vannak az oszlopnevek (kell, h legyen oszlopnév)
#ellenkező irányú '/'-jelre errort dob, erre figyelni

df = pandas.read_csv('g:/Saját meghajtó/08_phoenix/file_tries/Vevo_OPIT.csv', sep=';', header=0)
df.to_sql('asdtable', con=engine, if_exists='replace', index=False)
```

```python
#Column15 helyére beírni a potenciális kulcs oszlop nevét: 3 helyen kell átírni!
#lehet több potenciális kulcsoszlopot is megadni felsorolásként: Column14, Column15 etc.
#ha üres lista az eredmény, akkor megfelelő kulcsnak a megadott oszlop(ok)
#ha nem üres lista, akkor látjuk az oszlop(ok) ismétlődő értékét (/értékeit) és azt, hogy hányszor fordul elő ismétlődés
#pl: [(100783,2)] : ez azt jelenti, h a 100783-as érték kétszer is előfordul az adott oszlopban

engine.execute("WITH cte AS (SELECT Column15, ROW_NUMBER() OVER (PARTITION BY Column15 ORDER BY Column15) rownum FROM asdtable) SELECT * FROM cte WHERE rownum > 1;").fetchall()
```
